```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.style.available
```

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.style.use('dark_background') # 模板
plt.plot(x,y)
```

```python
plt.style.use('bmh')
plt.plot(x,y)
```

```python
plt.style.use('ggplot')
plt.plot(x,y)
```

```python
plt.style.use(['ggplot','bmh'])
plt.plot(x,y)
```

```python
# plt.xkcd() # 手绘风格 会影响全局 无法取消 除非重启kernel 
plt.xkcd()
plt.plot(x,y)
```
