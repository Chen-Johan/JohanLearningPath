```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# 设置字体以支持中文字符
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 生成三个不同标准差的正态分布数据
tang_data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# 创建一个8x6英寸的图形对象
fig = plt.figure(figsize=(8, 6))

# 绘制箱线图
plt.boxplot(tang_data, notch=False, sym='s', vert=True)

# 设置x轴刻度标签
plt.xticks([y + 1 for y in range(len(tang_data))], ['x1', 'x2', 'x3'])

# 设置x轴标签
plt.xlabel('x')

# 设置图表标题
plt.title('箱线图')

# 显示图表
plt.show()
           
```

```python
# 设置箱线图的所有组件颜色为黑色

# 生成三个不同标准差的正态分布数据
tang_data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# 创建一个8x6英寸的图形对象
fig = plt.figure(figsize=(8, 6))

# 绘制箱线图
bplot = plt.boxplot(tang_data, notch=False, sym='s', vert=True)

# 设置x轴刻度标签
plt.xticks([y + 1 for y in range(len(tang_data))], ['x1', 'x2', 'x3'])

# 设置x轴标签
plt.xlabel('x')

# 设置图表标题
plt.title('箱线图')

# 将箱线图的所有组件颜色设置为黑色
for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('black')
```

```python
# 生成三个不同标准差的正态分布数据
tang_data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# 创建一个8x6英寸的图形对象
fig = plt.figure(figsize=(8, 6))

# 绘制箱线图，notch=False表示不使用凹口，sym='s'表示异常值使用's'符号，vert=True表示垂直放置箱线图，patch_artist=True表示填充箱体颜色
bplot = plt.boxplot(tang_data, notch=False, sym='s', vert=True, patch_artist=True)

# 设置x轴刻度标签
plt.xticks([y + 1 for y in range(len(tang_data))], ['x1', 'x2', 'x3'])

# 设置x轴标签
plt.xlabel('x')

# 设置图表标题
plt.title('箱线图')

# 定义箱体颜色
colors = ['pink', 'lightblue', 'lightgreen']

# 为每个箱体设置颜色
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
```

# 小提琴图violinplot

```python
# 创建一个包含两个子图的图形对象，大小为12x5英寸
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# 生成四组不同标准差的正态分布数据
tang_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

# 在第一个子图中绘制小提琴图，不显示均值，显示中位数
axes[0].violinplot(tang_data, showmeans=False, showmedians=True)
axes[0].set_title('小提琴图')

# 在第二个子图中绘制箱线图
axes[1].boxplot(tang_data)
axes[1].set_title('箱线图')

# 为每个子图设置网格线和x轴刻度
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(tang_data))])

# 设置所有子图的x轴刻度标签
plt.setp(axes, xticks=[y + 1 for y in range(len(tang_data))], xticklabels=['x1', 'x2', 'x3', 'x4'])
```
