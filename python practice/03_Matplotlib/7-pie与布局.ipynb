```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 定义男性和女性的数量
m = 51212.
f = 40742.

# 计算男性和女性的比例
m_perc = m / (m + f)
f_perc = f / (m + f)

# 定义饼图的颜色和标签
colors = ['navy', 'lightcoral']
labels = ["Male", "Female"]

# 设置图形大小
plt.figure(figsize=(8, 8))

# 绘制饼图
paches, texts, autotexts = plt.pie([m_perc, f_perc], labels=labels, autopct='%1.1f%%', explode=[0, 0.05], colors=colors)

# 设置文本字体大小
for text in texts + autotexts:
    text.set_fontsize(20)

# 设置自动文本颜色为白色
for text in autotexts:
    text.set_color('white')
```

```python
import numpy as np

# 生成从0到10的1000个等间距的点
x = np.linspace(0, 10, 1000)

# 计算y1和y2的值
y2 = np.sin(x**2)
y1 = x**2

# 创建一个图形和一个子图
fig, ax1 = plt.subplots()

# 定义嵌入子图的位置和大小
left, bottom, width, height = [0.22, 0.45, 0.3, 0.35]
ax2 = fig.add_axes([left, bottom, width, height])

# 在主图中绘制y1
ax1.plot(x, y1)

# 在嵌入子图中绘制y2
ax2.plot(x, y2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# 定义一个函数，用于在条形图上添加标签
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax1.text(rect.get_x() + rect.get_width()/2., 1.02*height,
        "{:,}".format(float(height)),
        ha='center', va='bottom', fontsize=18)

# 定义前10个到达国家的名称和对应的值
top10_arrivals_countries = ['CANADA', 'MEXICO', 'UNITED\nKINGDOM', 'JAPAN', 
                            'CHINA', 'GERMANY', 'SOUTH\nKOREA', 'FRANCE', 'BRAZIL', 'AUSTRALIA']
top10_arrivals_values = [16.625687, 15.378026, 3.934508, 2.999718, 2.618737, 1.769498, 1.628563, 1.419409, 1.393710, 1.136974]

# 定义到达国家的名称和对应的百分比
arrivals_countries = ['WESTERN\nEUROPE', 'ASIA', 'SOUTH\nAMERICA', 'OCEANIA', 'CARIBBEAN', 
                      'MIDDLE\nEAST', 'CENTRAL\nAMERICA', 'EASTERN\nEUROPE', 'AFRICA']
arrivals_percent = [36.9, 30.4, 13.8, 4.4, 4.0, 3.6, 2.9, 2.6, 1.5]

# 创建一个图形和一个子图
fig, ax1 = plt.subplots(figsize=(20, 12))

# 绘制前10个到达国家的条形图
tang = ax1.bar(range(10), top10_arrivals_values, color='blue')

# 设置x轴的刻度标签
plt.xticks(range(10), top10_arrivals_countries, fontsize=18)

# 创建一个嵌入的子图
ax2 = inset_axes(ax1, width=6, height=6, loc=5)

# 定义饼图的爆炸效果
explode = (0.08, 0.08, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05)

# 绘制饼图
patches, texts, autotexts = ax2.pie(arrivals_percent, labels=arrivals_countries, autopct='%1.1f%%', explode=explode)

# 设置文本字体大小
for text in texts + autotexts:
    text.set_fontsize(16)

# 隐藏主图的边框
for spine in ax1.spines.values():
    spine.set_visible(False)

# 在条形图上添加标签
autolabel(tang)
```

```python
import numpy as np
from matplotlib.patches import Circle, Wedge, Polygon, Ellipse
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt

# 创建图形和子图
fig, ax = plt.subplots()

# 创建一个空列表用于存储所有的图形对象
patches = []

# 使用Wedge绘制完整的扇形和环形扇形
leftstripe = Wedge((.46, .5), .15, 90, 100)  # 默认绘制完整扇形
midstripe = Wedge((.5, .5), .15, 85, 95)                      
rightstripe = Wedge((.54, .5), .15, 80, 90)
lefteye = Wedge((.36, .46), .06, 0, 360, width=0.03)  # 当width < 1时绘制环形扇形
righteye = Wedge((.63, .46), .06, 0, 360, width=0.03)
nose = Wedge((.5, .32), .08, 75, 105, width=0.03)
mouthleft = Wedge((.44, .4), .08, 240, 320, width=0.01)
mouthright = Wedge((.56, .4), .08, 220, 300, width=0.01)
# 将所有扇形对象添加到patches列表中
patches += [leftstripe, midstripe, rightstripe, lefteye, righteye, nose, mouthleft, mouthright]

# 使用Circle绘制圆形
leftiris = Circle((.36, .46), 0.04)
rightiris = Circle((.63, .46), 0.04)
# 将所有圆形对象添加到patches列表中
patches += [leftiris, rightiris]

# 使用Polygon绘制多边形，通过传递顶点的坐标
leftear = Polygon([[.2, .6], [.3, .8], [.4, .64]], closed=True)
rightear = Polygon([[.6, .64], [.7, .8], [.8, .6]], closed=True)
topleftwhisker = Polygon([[.01, .4], [.18, .38], [.17, .42]], closed=True)
bottomleftwhisker = Polygon([[.01, .3], [.18, .32], [.2, .28]], closed=True)
toprightwhisker = Polygon([[.99, .41], [.82, .39], [.82, .43]], closed=True)
bottomrightwhisker = Polygon([[.99, .31], [.82, .33], [.81, .29]], closed=True)
# 将所有多边形对象添加到patches列表中
patches += [leftear, rightear, topleftwhisker, bottomleftwhisker, toprightwhisker, bottomrightwhisker]

# 使用Ellipse绘制椭圆
body = Ellipse((0.5, -0.18), 0.6, 0.8)
# 将椭圆对象添加到patches列表中
patches.append(body)

# 绘制所有的图形对象
colors = 100 * np.random.rand(len(patches))  # 设置随机颜色
p = PatchCollection(patches, alpha=0.4)
p.set_array(np.array(colors))
ax.add_collection(p)

# 显示图形
plt.show()
```
