最基本的一个图

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlable', fontsize = 16)
plt.ylabel('ylable')
```

不同的线条

字符|类型 | 字符|类型
---|--- | --- | ---
`  '-'	`| 实线 | `'--'`|	虚线
`'-.'`|	虚点线 | `':'`|	点线
`'.'`|	点 | `','`| 像素点
`'o'`	|圆点 | `'v'`|	下三角点
`'^'`|	上三角点 | `'<'`|	左三角点
`'>'`|	右三角点 | `'1'`|	下三叉点
`'2'`|	上三叉点 | `'3'`|	左三叉点
`'4'`|	右三叉点 | `'s'`|	正方点
`'p'`	| 五角点 | `'*'`|	星形点
`'h'`|	六边形点1 | `'H'`|	六边形点2 
`'+'`|	加号点 | `'x'`|	乘号点
`'D'`|	实心菱形点 | `'d'`|	瘦菱形点 
`'_'`|	横线点 | |

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

颜色

字符 | 颜色 color=''

```python
# 颜色 形状放一起
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro') # ro = red circle
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

绘制多个线

```python
tang_numpy = np.arange(0,10,0.5)
plt.plot(tang_numpy,tang_numpy,'r--')
plt.plot(tang_numpy,tang_numpy**2,'bs')
plt.plot(tang_numpy,tang_numpy**3,'go')
```

```python
plt.plot(tang_numpy,tang_numpy,'r--',
        tang_numpy,tang_numpy**2,'bs',
        tang_numpy,tang_numpy**3,'go')
```

指定线条的宽度

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth = 3.0) # 线宽
```

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
# linestyle 线型 marker 标记 markerfacecolor 标记颜色 markersize 标记大小
```

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth = 2.0, alpha = 0.4)
# setp 设置线的属性 color 颜色 linewidth 线宽 alpha 透明度
```

给图上加上注释

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('tang yu di:---')
plt.text(0,0,'tang yu di') # 在0,0的位置添加文字
plt.grid(True) # 添加网格

# 添加注释 xy 表示箭头的位置 xytext 表示文字的位置 arrowprops 表示箭头的属性 facecolor 箭头颜色 shrink 箭头的长度 headlength 箭头的长度 headwidth 箭头的宽度
plt.annotate('tangyudi',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='red',shrink=0.05,headlength= 20,headwidth = 20))
```

```python
x = range(10)
y = range(10)

fig = plt.gca() # 获取当前的图形 gca = get current axis 获取当前的轴
plt.plot(x,y)

fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

```python
import math
# 随机创建一些数据
x = np.random.normal(loc = 0.0,scale=1.0,size=300) 
width = 0.5
bins  =  np.arange(math.floor(x.min())-width,math.ceil(x.
max())+width,width)
ax = plt.subplot(111)

# 去掉上方和右方的坐标轴线
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
# 可以自己选择隐藏坐标轴上的锯齿线
plt.tick_params(bottom='off',top='off',left = 'off',right='off') 
# 加入网络
plt.grid() 
# 绘制直方图
plt.hist(x,alpha = 0.5,bins = bins)
```

```python
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = '10'
```

```python
x = range(10)
y = range(10)

# 创建标签列表
labels = ['tangyudi' for i in range(10)]

# 创建一个包含子图的图形
fig, ax = plt.subplots()

# 绘制x和y的折线图
plt.plot(x, y)

# 设置图形标题
plt.title('tangyudi')

# 设置x轴的刻度
ax.set_xticks(x)

# 设置x轴的刻度标签，并旋转45度，水平对齐方式为右对齐
ax.set_xticklabels(labels, rotation=45, horizontalalignment='right')
```
