```python
import matplotlib.pyplot as plt
import numpy as np
# from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# xy-4433
x = np.linspace(-4, 4, 33)  # 
y = np.linspace(-4, 4, 33)  # 

# 
X, Y = np.meshgrid(x, y)

# Zsin
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

# 
plt.show()
```

```python
# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# 3D
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

# z-2
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='rainbow')

# z
ax.set_zlim(-2, 2)

# 
plt.show()
```

```python
# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# theta-44100
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)

# z-22100
z = np.linspace(-2, 2, 100)

# rr = z^2 + 1
r = z**2 + 1

# xx = r * sin(theta)
x = r * np.sin(theta)

# yy = r * cos(theta)
y = r * np.cos(theta)

# 3D
ax.plot(x, y, z)

# 
plt.show()
```

```python
# 
np.random.seed(1)

# 
def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin

# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# 
n = 100

# z
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', 'x', -30, -5)]:
    # x2332
    xs = randrange(n, 23, 32)
    # y0100
    ys = randrange(n, 0, 100)
    # zzlowzhigh
    zs = randrange(n, int(zlow), int(zhigh))
    # 3D
    ax.scatter(xs, ys, zs, c=c, marker=m)

# 
plt.show()
```

```python
# 
np.random.seed(1)

# 
def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin

# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# 
n = 100

# z
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', 'x', -30, -5)]:
    # x2332
    xs = randrange(n, 23, 32)
    # y0100
    ys = randrange(n, 0, 100)
    # zzlowzhigh
    zs = randrange(n, int(zlow), int(zhigh))
    # 3D
    ax.scatter(xs, ys, zs, c=c, marker=m)

# 
ax.view_init(40, 0)

# 
plt.show()
```

```python
# 
fig = plt.figure()

# 3D
ax = fig.add_subplot(111, projection='3d')

# z3D
for c, z in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0]):
    # x019
    xs = np.arange(20)
    # y20
    ys = np.random.rand(20)
    # xs
    cs = [c] * len(xs)
    # 3D
    ax.bar(xs, ys, zs=z, zdir='y', color=cs, alpha=0.5)

# 
plt.show()
```
