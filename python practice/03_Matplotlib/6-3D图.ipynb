```python
import matplotlib.pyplot as plt
import numpy as np
# from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# 创建一个新的图形对象
fig = plt.figure()

# 创建一个3D坐标轴对象
ax = fig.add_subplot(111, projection='3d')

# 生成x轴和y轴的值，范围从-4到4，共33个点
x = np.linspace(-4, 4, 33)  # 确保第三个参数是整数
y = np.linspace(-4, 4, 33)  # 确保第三个参数是整数

# 生成网格点矩阵
X, Y = np.meshgrid(x, y)

# 计算Z轴的值，使用sin函数
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制3D曲面图，使用彩虹色映射
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

# 显示图形
plt.show()
```

```python
# 创建一个新的图形对象
fig = plt.figure()

# 创建一个3D坐标轴对象
ax = fig.add_subplot(111, projection='3d')

# 生成theta值，从-4π到4π，共100个点
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)

# 生成z轴的值，从-2到2，共100个点
z = np.linspace(-2, 2, 100)

# 计算半径r的值，r = z^2 + 1
r = z**2 + 1

# 计算x轴的值，x = r * sin(theta)
x = r * np.sin(theta)

# 计算y轴的值，y = r * cos(theta)
y = r * np.cos(theta)

# 绘制3D螺旋线
ax.plot(x, y, z)

# 显示图形
plt.show()
```

```python
# 设置随机种子，以确保每次运行代码时生成相同的随机数
np.random.seed(1)

# 定义一个函数，用于生成指定范围内的随机数
def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin

# 创建一个新的图形对象
fig = plt.figure()

# 创建一个3D坐标轴对象
ax = fig.add_subplot(111, projection='3d')

# 定义要生成的随机点的数量
n = 100

# 使用不同的颜色、标记和z轴范围生成随机散点图
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', 'x', -30, -5)]:
    # 生成x轴的随机值，范围在23到32之间
    xs = randrange(n, 23, 32)
    # 生成y轴的随机值，范围在0到100之间
    ys = randrange(n, 0, 100)
    # 生成z轴的随机值，范围在zlow到zhigh之间
    zs = randrange(n, int(zlow), int(zhigh))
    # 绘制3D散点图
    ax.scatter(xs, ys, zs, c=c, marker=m)

# 显示图形
plt.show()
```

```python
# 设置随机种子，以确保每次运行代码时生成相同的随机数
np.random.seed(1)

# 定义一个函数，用于生成指定范围内的随机数
def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin

# 创建一个新的图形对象
fig = plt.figure()

# 创建一个3D坐标轴对象
ax = fig.add_subplot(111, projection='3d')

# 定义要生成的随机点的数量
n = 100

# 使用不同的颜色、标记和z轴范围生成随机散点图
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', 'x', -30, -5)]:
    # 生成x轴的随机值，范围在23到32之间
    xs = randrange(n, 23, 32)
    # 生成y轴的随机值，范围在0到100之间
    ys = randrange(n, 0, 100)
    # 生成z轴的随机值，范围在zlow到zhigh之间
    zs = randrange(n, int(zlow), int(zhigh))
    # 绘制3D散点图
    ax.scatter(xs, ys, zs, c=c, marker=m)

# 设置视角，参数分别为仰角和方位角
ax.view_init(40, 0)

# 显示图形
plt.show()
```

```python
# 创建一个新的图形对象
fig = plt.figure()

# 创建一个3D坐标轴对象
ax = fig.add_subplot(111, projection='3d')

# 使用不同的颜色和z轴偏移量绘制3D柱状图
for c, z in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0]):
    # 生成x轴的值，范围为0到19
    xs = np.arange(20)
    # 生成y轴的随机值，共20个点
    ys = np.random.rand(20)
    # 生成颜色列表，长度与xs相同
    cs = [c] * len(xs)
    # 绘制3D柱状图，设置颜色和透明度
    ax.bar(xs, ys, zs=z, zdir='y', color=cs, alpha=0.5)

# 显示图形
plt.show()
```
