## 风格设置

```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<img src="f1.png" alt="FAO" width="990">

```python
def sinplot(flip=1):
    # 生成从0到14的100个等间距的数值
    x = np.linspace(0, 14, 100)
    
    # 循环生成并绘制正弦曲线
    for i in range(1, 7):
        # 计算正弦值并绘制曲线
        # x + i * 0.5: 控制正弦曲线的相位偏移
        # (7 - i) * flip: 控制正弦曲线的振幅和翻转
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
# 设置Seaborn的风格为白色网格
sns.set_style("whitegrid")

# 生成随机数据，并在每列数据上添加一个递增的偏移量
# 数据的形状为 (20, 6)，即20行6列
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2

# 使用Seaborn绘制箱线图
# data: 输入的数据
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks") # ticks风格 用于强调数据
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
#f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```
