```python
%matplotlib inline
import pandas as pd
```

```python
import numpy as np

# 折线图
s = pd.Series(np.random.randn(10),index = np.arange(0,100,10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), # cumsum(0)按列累加 cumsum(1)按行累加
               index = np.arange(0, 100, 10), 
               columns = ['A', 'B', 'C', 'D'])
df.head()
```

```python
# 折线图
df.plot()
```

```python
import matplotlib.pyplot as plt

# 柱状图
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot(ax = axes[0],kind='bar') # kind='bar' 竖向柱状图
data.plot(ax = axes[1],kind='barh') # kind='barh' 横向柱状图
```

```python
df = pd.DataFrame(np.random.rand(6, 4), 
               index = ['one', 'two', 'three', 'four', 'five', 'six'], 
               columns = pd.Index(['A', 'B', 'C', 'D'], name = 'Genus'))
df.head()  
```

```python
# 柱状图
df.plot(kind='bar')
```

```python
macro = pd.read_csv('./data/macrodata.csv')
macro.head()
```

```python
data = macro[['quarter','realgdp','realcons']]
```

```python
# 散点图 
data.plot.scatter('quarter','realgdp') # x轴为quarter，y轴为realgdp
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(data, color='g', alpha=0.3) # 散点矩阵图 color='g'颜色为绿色，alpha=0.3透明度为0.3

# 对角线上是每个变量的直方图，非对角线上是两两变量的散点图
```
